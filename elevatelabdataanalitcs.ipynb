```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
data = [['Product','Sales','Month'],["Laptop","1200","January"],["Phone","800","January"],["Tablet","400","January"],["Laptop","1500","February"],["Phone","900","February"],["Tablet","500","February"]]
with open("sales.csv","w",newline="") as file:
  writer = csv.writer(file)
  writer.writerows(data)
  file.close()
```

```python
df = pd.read_csv("sales.csv")
print(df)
```

```python
# Analysing Datas
```

```python
sales_by_product = df.groupby("Product")["Sales"].sum()
print(sales_by_product)
```

```python
sales_by_month= df.groupby("Month")["Sales"].sum()
sales_by_month
```

```python
#Bar Chart
plt.figure(figsize=(8,5))
sales_by_product.plot(kind="bar")
plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sales_by_month.plot(kind="line",marker="o")
plt.title("Sales by Month")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
#Pie Chart
plt.figure(figsize=(8,5))
sales_by_product.plot(kind="pie",autopct="%1.1f%%")
plt.title("Sales by Product")
plt.ylabel("")
plt.show()
```

```python
#Final Conclusion
#Laptop generated highest sales
# January had the highest sales
# plots shows a steep decrement in the sales
# Pie Chart clearly shows which products dominates the sales share.
```
